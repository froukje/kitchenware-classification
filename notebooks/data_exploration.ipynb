# Setup

```python
import glob
import os
import  random
import pandas as pd
from PIL import Image
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import seaborn as sns
```

# Read Images

```python
path = "../data/images"
```

```python
images = []
for file in glob.glob(os.path.join(path, "*.jpg")):
    images.append(file)
```

```python
images[:5]
```

Sort list of images by file name

```python
images.sort()
```

```python
images[:5]
```

```python
print(f"Nr of images: {len(images)}")
```

Plot some random images

```python
imgs = random.choices(images, k=10)
fig, axes = plt.subplots(2, 5, figsize=(15,10))
for i, ax in enumerate(axes.flatten()):
    img = mpimg.imread(imgs[i])
    ax.imshow(img)
    ax.set_xticks([])
    ax.set_yticks([])
```
