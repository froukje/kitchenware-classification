# Setup

```python
import os
import glob
import random
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

# Read Predictions

```python
path = "../predictions"
file = "prediction_2022-11-26-15-03.csv"
df_preds = pd.read_csv(os.path.join(path, file))
df_preds.head()
```

```python
# read test images
imgs_path = "../data/test"
file_list = []
for file in glob.glob(os.path.join(imgs_path, "*.jpg")):
    file_list.append(file)
```

```python
# select 50 random images and plot
k = 50
imgs = random.choices(file_list, k=k)
ids = [int(file.split("/")[-1].split(".")[0]) for file in imgs]
```

```python
classes = {0: 'cup', 1: 'fork', 2: 'glass', 3: 'knife', 4: 'plate', 5: 'spoon'}
```

```python
fig, axis = plt.subplots(10, 5, figsize=(20, 20))
for i, ax in enumerate(axis.flatten()):
    img = mpimg.imread(imgs[i])
    ax.imshow(img)
    ax.set_xticks([])
    ax.set_yticks([])
    label = df_preds[df_preds['Id'] == ids[i]]['label'].values[0]
    ax.set_title(f'Id: {ids[i]} - label: {classes[label]}')
```
